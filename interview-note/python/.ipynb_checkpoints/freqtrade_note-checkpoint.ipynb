GC 与 运行平台的详细信息

python可能有多种解释器： JPython、CPython、PyPy、IronPython
          'CPython' (C implementation of Python),
          'IronPython' (.NET implementation of Python),
          'Jython' (Java implementation of Python),
          'PyPy' (Python implementation of Python).


关于GC，我们知道Cpython的GC主要由引用计数、标记清除、分代机制组成。

引用计数回标记对象的引用数量，当引用数量为0的时候，会实时回收该对象占用的内存。这与GC触发时机无关。

引用计数存在循环引用的问题，因此依靠标记清除方法来回收这部分内存。
那么什么时候触发GC呢？
是依靠分代回收机制。

默认值是
当新生代对象到700时，触发一次minorGC（扫描新生代，剩余对象加入中年代）
当中年代对象到10时，触发一次majorGC（扫描新生、中年代，剩余对象加入中年、老年代）
当老年代对象到10时，触发一次更全面majorGC（扫描所有对象）

默认值可能导致gc频繁，gc存在stw，所以减少频率一定程度上可以减少stw对实时程序的影响。

```python
def gc_set_threshold():
    if platform.python_implementation() == "CPython":
        gc.set_threshold(50_000, 500, 1000)

# 第一个参数（50,000）：当分配的对象数量达到此阈值时，触发一次次要GC（minor GC）。
# 第二个参数（500）：当分配的对象数量达到此阈值时，触发一次主要GC（major GC）。
# 第三个参数（1000）：当分配的对象数量达到此阈值时，触发一次更全面的主要GC。
```

```python


self.strategy: IStrategy = StrategyResolver.load_strategy(self.config)



class StrategyResolver():
    
    @staticmethod
    def load_strategy(config: Optional[Config] = None) -> IStrategy:
        strategy: IStrategy = StrategyResolver._load_strategy(
            strategy_name, config=config, extra_dir=config.get("strategy_path")
        )
        #...
        return strategy

    @staticmethod
    def _load_strategy(name: str,config: Config, extra_dir: Optional[str] = None) -> IStrategy:
        """
        根据path 和 name 推出abs_paths，然后load_object
        """
        strategy = StrategyResolver._load_object(
            paths=abs_paths,
            object_name=strategy_name,
            add_source=True,
            kwargs={"config": config},
        )
        return strategy


    @classmethod
    def _load_object(
        cls, paths: List[Path], *, object_name: str, add_source: bool = False, kwargs: Dict) -> Optional[Any]:
        """
        Try to load object from path list.
        """

        (module, module_path) = cls._search_object(directory=_path, object_name=object_name, add_source=add_source)
        return module(**kwargs) # (**kwargs 表示dict) 初始化一个具体策略类的对象


    @classmethod
    def _search_object(
        cls, directory: Path, *, object_name: str, add_source: bool = False) -> Union[Tuple[Any, Path], Tuple[None, None]]:
        """
        Search for the objectname in the given directory
        :param directory: relative or absolute directory path
        :param object_name: ClassName of the object to load
        :return: object class
        """
        obj = next(cls._get_valid_object(Path, object_name), None)

        if obj:
            obj[0].__file__ = str(entry)
            if add_source:
                obj[0].__source__ = obj[1]
            return (obj[0], module_path)
        return (None, None)


    @classmethod
    def _get_valid_object(
        cls, module_path: Path, object_name: Optional[str], enum_failed: bool = False) -> Iterator[Any]:

        import importlib.util
        import inspect

        module_name = module_path.stem or ""  #获取文件名
        spec = importlib.util.spec_from_file_location(module_name, str(module_path)) # 创建一个模块规范对象，该对象包含了导入模块所需的所有信息，包括模块的名字和位置。
        module = importlib.util.module_from_spec(spec) # 创建模块实例
        spec.loader.exec_module(module) # 执行模块代码

        # 接下来，可以使用模块中的内容了

        valid_objects_gen = (
                (obj, inspect.getsource(module))
                for name, obj in inspect.getmembers(module, inspect.isclass) # 获取模块中的所有类
                if (
                    (object_name is None or object_name == name)
                    and issubclass(obj, cls.object_type)
                    and obj is not cls.object_type
                    and obj.__module__ == module_name
                )
            )
            # The __module__ check ensures we only use strategies that are defined in this folder.
        return valid_objects_gen




        
```

```python
## 定义DB和Session
## 一个Session是一个数据库连接

from sqlalchemy import create_engine, inspect
from sqlalchemy.exc import NoSuchModuleError
from sqlalchemy.orm import scoped_session, sessionmaker
from sqlalchemy.pool import StaticPool

def init_db(db_url: str) -> None:

    kwargs: Dict[str, Any] = {}

    if db_url == "sqlite://":
        kwargs.update( # dict update函数用于将一个字典添加到另一个字典中
            {
                "poolclass": StaticPool,
            }
        )
    if db_url.startswith("sqlite://"):
        kwargs.update(
            {
                "connect_args": {"check_same_thread": False}, #允许使用多线程进行DB IO
            }
        )


    engine = create_engine(db_url, future=True, **kwargs) #创建engine

    # 创建session
    Trade.session = scoped_session(sessionmaker(bind=engine, autoflush=False), scopefunc=get_request_or_thread_id)
    Order.session = Trade.session
    PairLock.session = Trade.session
    _KeyValueStoreModel.session = Trade.session
    _CustomData.session = scoped_session(sessionmaker(bind=engine, autoflush=True), scopefunc=get_request_or_thread_id)


    # 创建表
    ModelBase.metadata.create_all(engine)

```

```python
## 定义表结构

## 例如trade



from sqlalchemy.orm import DeclarativeBase ## 所有表的基类

class ModelBase(DeclarativeBase):
    pass


class Trade(ModelBase):
    __tablename__ = "trades" ## 表名
    session: ClassVar[SessionType] ## 处理这个表使用的session


    id: Mapped[int] = mapped_column(Integer, primary_key=True)  ## mapped_column是这将字段名映射到列名
    exchange: Mapped[str] = mapped_column(String(25), nullable=False)  # type: ignore
    is_open: Mapped[bool] = mapped_column(nullable=False, default=True, index=True)  # type: ignore
    fee_open: Mapped[float] = mapped_column(Float(), nullable=False, default=0.0)  # type: ignore
    fee_open_cost: Mapped[Optional[float]] = mapped_column(Float(), nullable=True)  # type: ignore
    open_date: Mapped[datetime] = mapped_column(nullable=False, default=datetime.now)


    # 关系表，加载trade表的时候，可以关联到orders表
    orders: Mapped[List[Order]] = relationship("Order", order_by="Order.id", cascade="all, delete-orphan", lazy="selectin", innerjoin=True) 
    custom_data: Mapped[List[_CustomData]] = relationship("_CustomData", cascade="all, delete-orphan", lazy="raise")

    # 1. orders: Mapped[List[Order]]
    # orders 是属性的名称。
    # Mapped[List[Order]] 是类型注解，表示这个属性是一个映射到数据库表的列表，列表中的每个元素都是 Order 类型的实例。
    
    # 2. relationship("Order", ...)
    # relationship 是 SQLAlchemy ORM 中的一个函数，用于定义两个实体之间的关系。
    # "Order" 是关联的实体类的名称。
    
    # 3. order_by="Order.id"
    # 这个参数指定了关联的 Order 实体列表的排序方式。
    # "Order.id" 表示按照 Order 实体的 id 字段进行排序。
    
    # 4. cascade="all, delete-orphan"
    # cascade 参数定义了级联操作的行为。
    # "all, delete-orphan" 表示所有操作（包括保存、更新、删除）都会级联到关联的 Order 实体。
    # "delete-orphan" 表示如果一个 Order 实体不再被任何其他实体引用，它将被自动删除。
    
    # 5. lazy="selectin"
    # lazy 参数定义了加载关联实体的方式。
    # "selectin" 表示使用 SELECT IN 查询来加载关联的实体列表。这是一种延迟加载策略，只有在访问 orders 属性时才会执行查询。
    
    # 6. innerjoin=True
    # innerjoin 参数定义了在加载关联实体时是否使用内连接（INNER JOIN）。
    # True 表示在加载 orders 属性时，会使用内连接来同时加载关联的 Order 实体。


    @validates("enter_tag", "exit_reason") ##validate是一个验证器，用于修改或者报错对应字段的指
    def validate_string_len(self, key, value):
        max_len = getattr(self.__class__, key).prop.columns[0].type.length
        if value and len(value) > max_len:
            return value[:max_len]
        return value
    
    def delete(self) -> None:
        for order in self.orders:
            Order.session.delete(order) # order的session和trade一样，提交trade commit即可

        Trade.session.delete(self) #删除
        Trade.commit()

    # session.add(new_trade) # 新增

    # trade_to_update = session.query(Trade).filter_by(id=1).first() #修改
    # user_to_update.is_open = fasle
    


    @staticmethod
    def commit():
        Trade.session.commit() #提交

    @staticmethod
    def rollback():
        Trade.session.rollback()#回滚


    # 查询
    @staticmethod
    def get_trades_proxy(
        *, # 后续使用key value参数
        pair: Optional[str] = None,
        is_open: Optional[bool] = None,
        open_date: Optional[datetime] = None,
        close_date: Optional[datetime] = None,
    ) -> List["LocalTrade"]:
        # 构建查询条件
        trade_filter = []
        if pair:
            trade_filter.append(Trade.pair == pair)
        if open_date:
            trade_filter.append(Trade.open_date > open_date)
        if close_date:
            trade_filter.append(Trade.close_date > close_date)
        if is_open is not None:
            trade_filter.append(Trade.is_open.is_(is_open))
            
        query = select(Trade).filter(*trade_filter) # *并非指针，固定写法
        query = query.options(lazyload(Trade.orders))
        
        return cast(List[LocalTrade], Trade.session.scalars(query).all()) #查询 + 类型转换

    ## 查询示例
    @staticmethod
    def get_overall_performance(minutes=None) -> List[Dict[str, Any]]:
        """
        Returns List of dicts containing all Trades, including profit and trade count
        NOTE: Not supported in Backtesting.
        """
        filters: List = [Trade.is_open.is_(False)]
        if minutes:
            start_date = datetime.now(timezone.utc) - timedelta(minutes=minutes)
            filters.append(Trade.close_date >= start_date)

        pair_rates = Trade.session.execute(
            select(
                Trade.pair,
                func.sum(Trade.close_profit).label("profit_sum"),
                func.sum(Trade.close_profit_abs).label("profit_sum_abs"),
                func.count(Trade.pair).label("count"),
            )
            .filter(*filters)
            .group_by(Trade.pair)
            .order_by(desc("profit_sum_abs"))
        ).all()

        return [
            {
                "pair": pair,
                "profit_ratio": profit,
                "profit": round(profit * 100, 2),  # Compatibility mode
                "profit_pct": round(profit * 100, 2),
                "profit_abs": profit_abs,
                "count": count,
            }
            for pair, profit, profit_abs, count in pair_rates
        ]

    @staticmethod
    def get_enter_tag_performance(pair: Optional[str]) -> List[Dict[str, Any]]:
        """
        Returns List of dicts containing all Trades, based on buy tag performance
        Can either be average for all pairs or a specific pair provided
        NOTE: Not supported in Backtesting.
        """

        filters: List = [Trade.is_open.is_(False)]
        if pair is not None:
            filters.append(Trade.pair == pair)

        enter_tag_perf = Trade.session.execute(
            select(
                Trade.enter_tag,
                func.sum(Trade.close_profit).label("profit_sum"),
                func.sum(Trade.close_profit_abs).label("profit_sum_abs"),
                func.count(Trade.pair).label("count"),
            )
            .filter(*filters)
            .group_by(Trade.enter_tag)
            .order_by(desc("profit_sum_abs"))
        ).all()

        return [
            {
                "enter_tag": enter_tag if enter_tag is not None else "Other",
                "profit_ratio": profit,
                "profit_pct": round(profit * 100, 2),
                "profit_abs": profit_abs,
                "count": count,
            }
            for enter_tag, profit, profit_abs, count in enter_tag_perf
        ]

    @staticmethod
    def get_exit_reason_performance(pair: Optional[str]) -> List[Dict[str, Any]]:
        """
        Returns List of dicts containing all Trades, based on exit reason performance
        Can either be average for all pairs or a specific pair provided
        NOTE: Not supported in Backtesting.
        """

        filters: List = [Trade.is_open.is_(False)]
        if pair is not None:
            filters.append(Trade.pair == pair)
        sell_tag_perf = Trade.session.execute(
            select(
                Trade.exit_reason,
                func.sum(Trade.close_profit).label("profit_sum"),
                func.sum(Trade.close_profit_abs).label("profit_sum_abs"),
                func.count(Trade.pair).label("count"),
            )
            .filter(*filters)
            .group_by(Trade.exit_reason)
            .order_by(desc("profit_sum_abs"))
        ).all()

        return [
            {
                "exit_reason": exit_reason if exit_reason is not None else "Other",
                "profit_ratio": profit,
                "profit_pct": round(profit * 100, 2),
                "profit_abs": profit_abs,
                "count": count,
            }
            for exit_reason, profit, profit_abs, count in sell_tag_perf
        ]




```

```python
def _init(self) -> None:
    asyncio.run(self._startup_telegram())


async def _startup_telegram(self) -> None:
    await self._app.initialize()
    await self._app.start()
    if self._app.updater:
        await self._app.updater.start_polling(
            bootstrap_retries=-1,
            timeout=20,
            # read_latency=60,  # Assumed transmission latency
            drop_pending_updates=True,
            # stop_signals=[],  # Necessary as we don't run on the main thread
        )
        while True:
            await asyncio.sleep(10)
            if not self._app.updater.running:
                break
```

```python
with关键字 自定义时间统计：

实现类中包含 __enter__ 和 __exit即可使用with。

在程序段执行时调用enter，执行完毕后调用exit
```

```python
class MeasureTime:
    """
    Measure the time of a block of code and call a callback if the time limit is exceeded.
    """

    def __init__(
        self, callback: Callable[[float, float], None], time_limit: float, ttl: int = 3600 * 4
    ):
        self._callback = callback
        self._time_limit = time_limit
        self.__cache: TTLCache = TTLCache(maxsize=1, ttl=ttl)

    def __enter__(self):
        self._start = time.time()

    def __exit__(self, *args):
        end = time.time()
        if self.__cache.get("value"):
            return
        duration = end - self._start

        if duration < self._time_limit:
            return
        self._callback(duration, self._time_limit)

        self.__cache["value"] = True
```
